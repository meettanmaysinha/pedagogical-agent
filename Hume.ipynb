```python
import json
 
# Opening JSON file
f = open('predictions.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
```

```python
data
```

```python
results = data[0]["results"]["predictions"][0]["models"]["face"]["grouped_predictions"]
```

```python

results
```

```python
lol = 0
```

```python
# Breakdown by ID
results_by_id = pd.DataFrame()
id_results_clean = pd.DataFrame()
for id_results in results:
    face_id = id_results["id"]
    id_results_clean["predictions"] = id_results["predictions"]
    id_results_clean["id"] = face_id
    results_by_id = pd.concat([results_by_id,id_results_clean],ignore_index=True)
```

```python
results_by_id["predictions"]
```

```python
extracted_results = pd.json_normalize(results_by_id["predictions"])
extracted_results["id"] = results_by_id["id"]
extracted_results
```

```python
def sort_emotions(emotions):
    # Sort emotions by scores
    emotions.sort(key=lambda x: x['score'],reverse=True)
    emotions
```

```python
extracted_results.apply(lambda x: sort_emotions(x["emotions"]),axis=1)
```

```python
extracted_results
```

```python
extracted_results["top3_emotions"] = extracted_results.apply(lambda x: x["emotions"][:3],axis=1)
```

```python
extracted_results["emotion1"] = extracted_results.apply(lambda x: x["emotions"][0]["name"],axis=1)
extracted_results["emotion1_score"] = extracted_results.apply(lambda x: x["emotions"][0]["score"],axis=1)
extracted_results["emotion2"] = extracted_results.apply(lambda x: x["emotions"][1]["name"],axis=1)
extracted_results["emotion2_score"] = extracted_results.apply(lambda x: x["emotions"][1]["score"],axis=1)
extracted_results["emotion3"] = extracted_results.apply(lambda x: x["emotions"][2]["name"],axis=1)
extracted_results["emotion3_score"] = extracted_results.apply(lambda x: x["emotions"][2]["score"],axis=1)
```
